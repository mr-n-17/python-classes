```python
list1length=int(input("ENTER THE LENGHTH OF LSIT 1"))
list2length=int(input("ENTER THE LENGHTH OF LSIT 2"))
l1=[]
l2=[]
for i in range(list1length):    
    l1.append(int(input(f'ENTER THE INPUT {i+1}')))
for i in range(list2length):    
    l2.append(int(input(f'ENTER THE INPUT {i+1}')))
print(l1)
print(l2)
l3=[]
l4=[]
c=min(list1length,list2length)
for i in range(c):
    if c==len(l2) and l2[i] in l1:
        l3.append(l2[i])
    elif c==len(l1) and l1[i] in l2:
        l3.append(l1[i]) 
    elif c==len(l2) and l2[i] not in l1:
        l4.append(l2[i])
    elif c==len(l1) and l1[i] not in l2:
        l4.append(l1[i]) 
print("LIST OF COMMON ELEMENTS IN L1 and L2 ",l3)
print("LIST OF UNIQUE ELEMENTS IN L1 AND L2 ",l4)
print("SORTED OF LIST3 IS : ",sorted(l3))
print("SORTED OF LIST4 IS : ",sorted(l4,reverse=True))

    
```

```python
def zippingit(a,b,c):
    x=list(zip(a,b,c))
    dict1={}
    for i in x:
        dict1[i[0]]=[*i[1:]]
    return dict1

        
Headings = ['first_name', 'last_name', 'age'] 
Person1 = ['Mark','Taylor',20] 
Person2 = ['Emma','Jones',30]
print(zippingit(Headings,Person1,Person2))
```

```python
class Bankaccount:
    def __init__(self):
        self.balance=0

    def deposit(self,amount):
        self.balance+=amount 
        return self.balance 

    def withdraw(self,amount):
        if self.balance >= amount :
            self.balance-=amount 
            return self.balance
        return "YOU DON'T HAVE ENOUGH BALANCE"

sam=Bankaccount()
rio=Bankaccount()
print(sam.deposit(100))
print(rio.deposit(200))
print(sam.withdraw(20))
print(rio.withdraw(30))

        
```
